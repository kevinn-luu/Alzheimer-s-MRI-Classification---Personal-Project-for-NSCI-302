The proportion of people with Alzheimer's has increased over the years due to the growing number of elderly people in the general population. Interestingly, during the classification of this data, models may detect very small nuances and features missed by human classification that can be reliably used in diagnoses. Berl et al. (2023) demonstrated that a CNN could classify sex from retinal images, even without large training datasets. In their study, they also visualized the saliency of their trained model, identifying the areas of the images that contributed to the model's classification decisions and finding reliable differences between sexes in the retina.

In this project, a CNN is used to classify MRI images of brains. The model will predict whether or not a given brain image shows Alzheimer's disease or not.

## Transforming data
Images need to changed into a proper format in order to be passed to the machine learning model. The function below is one that can be applied to the dataset to decode, resize, and normalize images. The machine learning mode also expects images to be inputted as arrays. The architecture of the model (refer to model_architecture.pdf) expects images to be of shape (256, 256, 3). So images are 256px x 256px and are RGB coloured.
### References
- __[Numpy stack](https://numpy.org/doc/stable/reference/generated/numpy.stack.html)__

```python
def decode_and_preprocess_data(pq_img):
    decoded_image = Image.open(io.BytesIO(pq_img['bytes'])) #decode image
    decoded_image = decoded_image.resize((256, 256)) #resize dimensions to 256 x 256
    decoded_image_array = np.array(decoded_image) #turn image into an array 
    

    
    #the model expects the shape of the image to be (256, 256, 3). The "3" refers to the color channels of the image
    if decoded_image_array.shape[-1] != 3 or len(decoded_image_array) == 2: #check if the image has the proper shape
        #if not, convert the grayscale image into a new image by stacking 2D arrays along the last axis to make a new 3D array to produce coloured images from grayscale
        #https://numpy.org/doc/stable/reference/generated/numpy.stack.html
        decoded_image_array = np.stack((decoded_image_array,)*3, axis=-1) 
    #normalize pixel values of image to be [0, 1]    
    decoded_image_array = decoded_image_array/255.0
    #return type as float32 to reduce the amount of memory used
    return decoded_image_array.astype(np.float32)
```

```python
#drop the old label column to reduce dataset size and memory 
training_data = training_data.drop(['label'], axis = 1)
testing_data = testing_data.drop(['label'], axis = 1)
```

Below, the function defined is applied to the dataset to create a new column called "array". This column stores the image data as RGB images in array form with their pixel values normalized.

```python
training_data['array'] = training_data['image'].apply(decode_and_preprocess_data)
testing_data['array'] = testing_data['image'].apply(decode_and_preprocess_data)
```

## Training and testing data

```python
#Training and testing arrays 
#stack all the arrays into one array i.e. X_train
X_train = np.stack(training_data['array'].values)
X_test = np.stack(testing_data['array'].values)

y_train = np.stack(training_data["Alzheimer's"].values)
y_test = np.stack(testing_data["Alzheimer's"].values)
```

Here we verify that the shape of the training and testing data makes sense. The output below means that in the training set, there are 5120 total images each one having dimension 256 x 256 and they are colored.

```python
print(X_train.shape)
print(y_train.shape)
```

Data will be passed to the machine learning model as Dataset objects from the tensorflow library. Doing this simply makes the training process much more efficient. Tensorflow datasets only load the data into memory when that data is needed during training. This is especially useful for larger datasets that may not all fit into memory. Dataset objects also allow for the piping of operations. Multiple operations can be done in sequence on a Dataset which allows for easy and efficient data processing.

```python
#produce tensorflow Dataset objects
model_training_dataset = tf.data.Dataset.from_tensor_slices((X_train, y_train))
model_testing_dataset = tf.data.Dataset.from_tensor_slices((X_test, y_test))
```

```python
model_training_dataset
```

Data is usually fed to machine learning models in batches. The model goes through one batch of data at a time instead of seeing all images at once. For this model, a batch_size of 32 is used. Since there 5120 images in totdal, batching the data with batch size 32 would produce 160 epochs. The data is also shuffled which prevents the model from seeing too much of the same type of image during rounds of training.

```python
BATCH_SIZE = 32

#https://www.tensorflow.org/guide/data_performance
model_training_dataset = model_training_dataset.shuffle(len(X_train)).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
model_testing_dataset = model_testing_dataset.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
```

# Model Architecture
A Convoluted Neural Network (CNN) is used to classify the image data. CNNs very commonly used in image classification problems. Using a CNN for image classification has many benefits compared to using other types of neural networks (NN) for these types of problems. In a CNN, a neuron doesn't have to be connected to every single other neuron in the previous layer. It creates more local connections instead. This speeds up the training process by reducing the weights inputted to the next layer. CNNs also make use of sliding filters. The size of these filters can be specified when building the model. Each convolutional layer can have its own filter. Silding filters move across the data and examine multiple values at once. Imagine a box moving across an image. The model analyzes the information contained in the box, then moves the box to the next location and does so again. The sliding filter allows CNNs to capture nuances and recognize patterns contained in images.

```python
#import libraries for model building and training
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Input
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.callbacks import EarlyStopping
```

In CNNs, each convoluted layer is usually followed by a pooling layer. A convoluted layer passes a feature map to the next layer. Pooling layers help reduce the dimensions of this map which decreases the training time. The MaxPooling2D layer extracts the most important feature from the map - the one with the highest value. This allows the model to only focus on the most important information and not consider other features that may not contribute much to model learning.

Before the final Dense layer, there is a Flatten layer. This is necessary because Convoluted layers work on multidimensional feature maps; however, in this case, the model uses a one-dimensional vector to make its prediction. The flatten layer preserves all the important spatial features of the image learned by the model and then reduces the size of those feature maps to one dimension.

A dropout layer is used to help the model prevent overfitting. In the model below, Dropout(0.5) is used. This means that in a training round, every neuron in the model has a 50% chance of being dropped from the model in that round. As a result, the remaining neurons scale up their outputs, in this case, by 2 so that the overall distribution of outputs and weights remains the same.

```python
#Model architecture
model = Sequential([
    Input(shape=(256, 256, 3)),
    Conv2D(32, (3, 3), activation='relu'),
    MaxPooling2D(),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D(),
    Flatten(),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])
```

```python
model.summary()
```

```python
#Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

# Model Training
Below, the model is trained with 20 epochs meaning that the 160 batches of data will be seen 20 times. To prevent overfitting, early stopping is used. Early stopping monitors the loss on the validation set. In this case, if the validation loss experiences no significant decrease for three consecutive training rounds, the training process is stopped and the previous best weights are used.

```python
#Training the model
early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
history = model.fit(model_training_dataset, validation_data=model_testing_dataset, epochs=20, callbacks=[early_stop])
```

Below, accuracy and loss is plotted for both training and validation sets. This data is acessed from the model history which was previously stored in a variable. The plots visualize how accuracy and loss changed through the 20 epochs of training. A well-trained model should have accuracy increase and loss decreasing. The validation curves should follow this pattern too.

```python
#Plot loss and accuracy over training
plt.figure(figsize = (20, 10))

#accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title("Accuracy")
plt.legend()

#loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title("Loss")
plt.legend()

plt.tight_layout()
plt.show()
```

# Model Testing
The above plots suggest that the model is performing well and has not overfit the training data. Now, the model can be used on the test set to further evaluate its performance.

```python
#Evaluate on the test set 
test_loss, test_accuracy = model.evaluate(model_testing_dataset)
print(test_loss)
print(test_accuracy)
```

Evaluation on the test set produces high accuracy and low loss. To actually see the model predictions, the below code produces the predictions and the actual labels from one batch of data. The model will output its predictions as a probability i.e. a prediction of 0.87 would mean that the model is 87% sure that a particular image belongs to a particular class. 

```python
#For one batch, produce actual labels and predicted labels
for images, labels in model_testing_dataset.take(1):
    model_predictions = model.predict(images) #predict on batch of images
    predicted_labels = (model_predictions > 0.5).astype(int) #convert predicted probabilities to integers
#print predicted labels and actual labels
print("Predicted labels:", predicted_labels.flatten())
print("Actual labels:", labels.numpy())
```

# Saliency Map

### References
- __[GradientTape for Keras Users](https://medium.com/analytics-vidhya/tf-gradienttape-explained-for-keras-users-cc3f06276f22)__
- ChatGPT 
    - prompt and code generated indicated below

```python
#Functions for producing saliency maps
def compute_saliency_map(model, image, class_index):
    '''
    Computes the saliency map for a single image and its specified clas
    Args:
        model: trained CNN model
        image: input image
        class_index: binary classification for the input image (1 or 0)
    
    Returns:
        saliency map for the given image
    '''
    
    image_tensor = tf.convert_to_tensor(image[np.newaxis, ...])
    
    #https://medium.com/analytics-vidhya/tf-gradienttape-explained-for-keras-users-cc3f06276f22
    with tf.GradientTape() as tape:
        tape.watch(image_tensor)
        predictions = model(image_tensor)
        
        if class_index == 1:
            loss = predictions[:, 0]
        else:
            loss = 1 - predictions[:, 0]
            
    gradients = tape.gradient(loss, image_tensor)
    saliency = tf.reduce_max(tf.abs(gradients), axis=-1)[0]
        
    saliency = (saliency - tf.reduce_min(saliency)) / (tf.reduce_max(saliency) - tf.reduce_min(saliency) + 1e-8)
    return saliency.numpy()

def visualize_saliency(image, saliency_map):
    '''
    Overlays saliency_map on image for visualization
    '''
    
    image = (image*255).astype("uint8")
    
    plt.figure(figsize=(10,5))
    
    plt.subplot(1, 2, 1)
    plt.imshow(image.astype("uint8"))
    plt.title("Original Image")
    plt.axis("off")
    
    plt.subplot(1, 2, 2)
    plt.imshow(image.astype("uint8"), alpha=0.6)
    plt.imshow(saliency_map, cmap='hot', alpha=0.4)
    plt.title("Saliency Map")
    plt.axis("off")
    
    plt.show()
```

```python
#Visualize saliency

image_iterator = model_testing_dataset.as_numpy_iterator()
one_batch = image_iterator.next()

image = one_batch[0][0]

saliency = compute_saliency_map(model, image, 1)

visualize_saliency(image, saliency)
```
